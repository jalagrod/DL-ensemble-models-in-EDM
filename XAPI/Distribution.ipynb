```python
import pandas as pd
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt  
import seaborn as sns 
from scipy.stats import shapiro, kstest, anderson, normaltest, probplot
```

```python
DATA_PATH = "xapi.csv"
TARGET = "Class"

# Load dataset
df = pd.read_csv(DATA_PATH)
```

```python
print(df.columns)
```

```python
# Remove spaces in column names (if any)
df.columns = df.columns.str.replace(' ', '')

# Convert categorical variables using One-Hot Encoding (for nominal variables)
df = pd.get_dummies(df, drop_first=True)

# Convert boolean values (if any)
df = df.applymap(lambda x: 1 if x == True else (0 if x == False else x))

# Convert all remaining non-numeric variables using Ordinal Encoding
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category').cat.codes
```

```python
print(df.columns)
```

```python
features = df.select_dtypes(include=['number']).columns
```

```python
for feature in features:
    stat, p_value = shapiro(df[feature])  # Shapiro-Wilk Test
    print(f'Shapiro-Wilk Test for {feature}: Statistic={stat:.4f}, p-value={p_value:.4f}')
    
    stat, p_value = normaltest(df[feature])  # D'Agostino-Pearson Test
    print(f'D\'Agostino-Pearson Test for {feature}: Statistic={stat:.4f}, p-value={p_value:.4f}\n')
```
